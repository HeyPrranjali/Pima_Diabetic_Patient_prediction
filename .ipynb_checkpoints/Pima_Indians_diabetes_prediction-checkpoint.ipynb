## About the Dataset

* __Predictor Features(Independent features)__<br>
1. Pregnancies: Number of time patient has been pregnant<br>
2. Glucose: Plasma glucose concentration<br>
3. Blood Pressure: Diastolic blood pressure<br>
4. Skin Thickness: Tricept skinfold thickness<br>
5. Insuline: 2 hrs serum insulin<br>
6. BMI: Body mass index<br>
7. Diabetes Pedigree Function: Family history of Diabetes<br>
8. Age: Age of person in years<br>
* __Target Feature (Dependent/Output Feature)__<br>
9. Outcome: 0 (the person is Non-Diabetic), 1(person is Diabetic)


### Medical reference table:<br>
| Features              | Normal Range(medical context) | Notes and Interpretation |
| :---------------- | :------ | :---- |
| Pregnancies       |   0-10 (Typically)   | not a medical range but >6 is considered high parity |
|||||
| Glucose(mg/dl)          |   Fasting: 70-99| Glucose ≥ 200 mg/dL in OGTT indicates diabetes. This dataset uses OGTT-like measurement. |
|                         |  Pre-Diabetic: 100-125     |OGTT: Oral Glucose Tolerance Test |
|                         |  Diabetic: >=126     | |
|||||
| BloodPressure (mm Hg)   |  Normal: <120/80  | Diastolic BP < 60 is low; >90 is considered high (hypertension).|
|                         |  Diastolic (used here): 60–80     | |
|||||
| SkinThickness (mm) |  Typical: 10–40 mm | Measures subcutaneous fat. Zero or very high values are likely errors or outliers.|
|||||
| Insulin (μU/mL)                |  Normal fasting insulin: 2–25 μU/mL     | In this dataset, a 2-hour insulin after glucose load is expected ~16–166 μU/mL. |
|||||
| BMI (kg/m²)                         | Normal: 18.5–24.9    |Major risk factor for type 2 diabetes. |
|                         |  Overweight: 25–29.9    | |
|                         |  Obese: ≥30     | |
|||||
|DiabetesPedigreeFunction | No strict clinical scale; usually 0.1 – 2.5  | Higher = greater hereditary risk. 1.0+ is usually considered significant. |
|||||
| Age (years)                      |  Dataset: 21–81, Risk increases after age 45  | Age is a known risk factor for type 2 diabetes.|
> Zero values in features like Glucose, BMI, Insulin, etc. can be physiologically impossible, indicating missing data.<br>
> These should be:<br>* either Replaced with NaN <br>* or Imputed using mean, median, or model-based methods

```python
# Numarically detect lower bound, upper bond and outlier counts for each feature
#create a function to detect the outier count
def count_outliers_for_all_features(df):
    outlier_summary= []
    outli= pd.DataFrame()
    
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        q1= df[col].quantile(0.25)
        q3= df[col].quantile(0.75)
        iqr= q3-q1
        lower_bound= q1- 1.5*(iqr)
        upper_bound= q3+ 1.5*(iqr)
        
        outliers= df[(df[col]<lower_bound) | (df[col]>upper_bound)]
        
        outlier_summary.append({
        'column':col,
        'lower bound':round(lower_bound,2),
        'upper bound':round(upper_bound,2),
        'outlier count':len(outliers)
        })
        
        if outliers.empty:
            outlier_rows= pd.concat([outlier_rows,outliers])
            
        outlier_rows= outliers.drop_duplicates()
        outlier_summary_df= pd.DataFrame(outlier_summary).sort_values(by="outlier count", ascending=False)
        
    print("outlier summary (by IQR method)")
    print(outlier_summary_df.to_string(index=False))
        
    return outlier_summary_df, outlier_rows


outlier_df= count_outliers_for_all_features(db_data)
print(len(outlier_df))
```

__What The Outlier Summary Tells Us:__
* Columns like DiabetesPedigreeFunction, Insulin, and BloodPressure have notable outliers.
* Glucose and Outcome have no outliers as per IQR.
* Negative lower bounds for columns like Age, Insulin, Pregnancies indicate possible zero or low value issues (should not be negative or zero in a real medical context).
* These are not "bad" data, just unusual compared to most of the data.

__After the Summary:__
* Document the Decisions and Always record:
* Which method you used (here IQR)
* How many records were impacted( here 29 for pedigree, 24 for Insulin, 14 for BP, 9 for Age, 8 for BMI, 4 for Pregnancy, 3 for SkinThikness)
* Whether you capped or removed(if data is < 5%	Usually acceptable and can be removed)
* but in healthcare datasets, we don’t remove outliers blindly. We use robust techniques like IQR capping or scaling, because edge cases may represent rare but critical patient profiles.

```python
def cap_outliers_iqr(df, features):
    df_capped = df.copy()
    for col in features:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_capped[col] = df[col].clip(lower, upper)
    return df_capped

# Columns to check for outliers
cols = db_data.columns.drop('Outcome')
df_capped = cap_outliers_iqr(db_data, cols)
print(df_capped)
```
